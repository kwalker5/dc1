# Read in each file using read csv

```python
import pandas as pd

test_x = pd.read_csv('test_x_region.csv')
train_x = pd.read_csv('train_x_region.csv')
train_y = pd.read_csv('train_y.csv')
```

# Feature Engineering

```python
corr = df.corr()
corr['Immigration to the United States'].sort_values(ascending = False, key = abs)
```

Showing the correlation of the colums of the data frame before doing any feature engineering

# Creating columns that I think will have good correlation

```python
df['Happiness per family'] = df['Happiness Score']/df['Family']
df['Happiness Rank By Health'] = df['Happiness Rank']/df['Health (Life Expectancy)']
df['Happiness by Trust'] = df['Happiness Rank']/df['Trust (Government Corruption)']
```

```python
corr = df.corr()
corr['Immigration to the United States'].sort_values(ascending = False, key = abs)
```


Of the columns that I created Happiness rank by health had the most correlation 

# Using label binarizer to turn the objects in regions to int values

```python
from sklearn.preprocessing import LabelBinarizer

enc = LabelBinarizer()
enc.fit(df['Region'])
transformed_op = enc.transform(df['Region'])
transformed_op
```

```python
one_hot_df = pd.DataFrame(transformed_op)
df = pd.concat([df, one_hot_df],axis=1)
df
```

```python
df.drop(['Region', 'Country'], axis = 1, inplace = True)
df
```

Added the int values from region in to the data frame and dropping the country column

# Testing the correlation after the feature engineering

```python
corr = df.corr()
corr['Immigration to the United States'].sort_values(ascending = False, key = abs)
```


The value 1 and population estimate have the strongest correlation to immigration to the united states

# Finding the best metric for the data set 

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(df.drop(['Immigration to the United States'],axis =1), train_y)
```

```python
reg.score(df.drop(['Immigration to the United States'],axis =1), train_y)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
params = {'alpha' : [1e-3,0.01,0.1,1,10,15,50, 100, 1000,100000,1000000]}
lasso = Lasso()
reg = GridSearchCV(lasso, params,scoring='neg_mean_squared_error',cv=5)
reg.fit(df.drop(['Immigration to the United States'],axis = 1), train_y)
```

```python
reg.best_score_
```

```python
reg.best_params_
```

```python
lasso = Lasso(alpha=100)
lasso.fit(df.drop(['Immigration to the United States'],axis =1), train_y)
lasso.score(df.drop(['Immigration to the United States'],axis =1), train_y)
```

```python
from sklearn.linear_model import Ridge
params = {'alpha' : [1e-3,0.01,0.1,1,10,15,50, 100, 1000,100000,1000000]}
ridge = Ridge()
reg = GridSearchCV(ridge, params,scoring='neg_mean_squared_error',cv=5)
reg.fit(df.drop(['Immigration to the United States'],axis = 1), train_y)
```

```python
reg.best_score_
```

```python
reg.best_params_
```

```python
ridge = Ridge(alpha=15)
ridge.fit(df.drop(['Immigration to the United States'],axis =1), train_y)
ridge.score(df.drop(['Immigration to the United States'],axis =1), train_y)
```
